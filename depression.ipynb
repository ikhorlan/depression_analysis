<h4>Logistic Regression<h4>

```python
X = b_dep_dc.drop(columns=["depressed"])
y = b_dep_dc['depressed']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
#X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, random_state=2)
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
X_scaled = scaler.transform(X_train)
```

```python
lr=LogisticRegression()

lr.fit(X_scaled, y_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
lr.score(X_test_scaled, y_test)
```

```python
lr.score(X_scaled, y_train)
```

```python
inspector = get_inspector(lr, X, y)
inspector.show_model()
```

<h4> Depression among students<h4>

```python
dep=pd.read_csv('../Datasets/depression_anxiety_data.csv')
```

```python
dep.head()
```

```python
dep.columns
```

```python
dep.head()
```

```python
dep.depressiveness.unique()
```

```python
dep.depression_severity.unique()
```

```python
depr=dep.dropna()
```

```python
dep_d=pd.DataFrame()
dcolumns=['gender']
for column in dcolumns:
    df_dummy=pd.get_dummies(depr[column], prefix=column, drop_first=True)
    dep_d=pd.concat([depr, df_dummy], axis=1)
    
print(dep_d.shape)
```

```python
keyword_mappings = {'None-minimal': 0, 'Mild': 1,'Moderate': 2, 'Moderately severe': 3, 'Severe': 5}

dep_d['dep_sev'] = dep_d.depression_severity.map(keyword_mappings)
```

```python
dep_dd=pd.DataFrame()
dcolumns=[[
    'depressiveness',
    'suicidal',
    'anxiousness', 
    'anxiety_diagnosis', 
    'anxiety_treatment', 
    'sleepiness', 
    'depression_diagnosis',
    'depression_treatment'
]]
for column in dcolumns:
    df_dummy=pd.get_dummies(dep_d[column], prefix=column, drop_first=True)
    dep_dd=pd.concat([dep_d, df_dummy], axis=1)
```

```python
dep_dd
```

```python
dep_clean = dep_dd.drop([
    'gender', 
    'who_bmi', 
    'depression_severity', 
    'depressiveness', 
    'suicidal', 
    'anxiousness',
    'anxiety_diagnosis', 
    'anxiety_treatment', 
    'sleepiness', 
    'depression_diagnosis', 
    'depression_treatment',
    'anxiety_severity'
], axis=1)
```

```python
dep_clean.head()
```

```python
dep_clean.corr()
```

```python
dep_clean_sev=dep_clean.groupby('dep_sev').agg({
    'depressiveness_True':['mean', 'sum'],
    'suicidal_True':['mean', 'sum'], 
    'depression_diagnosis_True':['mean', 'sum'],
    'depression_treatment_True': ['mean', 'sum']
    
})
```

```python
dep_clean_sev
```

```python
dep_clean_year=dep_clean.groupby('school_year').agg({
    'dep_sev':['mean', 'sum'],
    'depressiveness_True':['mean', 'sum'],
    'suicidal_True':['mean', 'sum'], 
    'depression_diagnosis_True':['mean', 'sum'],
    'depression_treatment_True': ['mean', 'sum']    
})
```

```python
dep_clean_year
```

```python
dep_clean_sev.columns
```

```python
dep_new=dep_clean_sev.rename(columns={
    ('depressiveness_True',   'sum'):'depressiveness',
    ('depression_diagnosis_True',   'sum'): 'diagnosis',
    ('depression_treatment_True',   'sum'): 'treatment'
}, inplace=True)
```
